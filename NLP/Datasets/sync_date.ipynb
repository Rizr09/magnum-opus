```python
import os
import pandas as pd
from pathlib import Path

def clean_title(title):
    # Remove '\r' characters and strip leading/trailing whitespace
    return title.replace('\r', '').strip()

def process_csv_files(directory):
    # Create an empty DataFrame to store the combined data
    df_combined = pd.DataFrame(columns=['Date', 'Title'])

    # Get the full path of the directory
    data_dir = Path(directory)

    # Iterate through all CSV files in the directory
    for file in data_dir.glob('*.csv'):
        # Read the CSV file
        df = pd.read_csv(file, sep=';', encoding='utf-8', lineterminator='\n')
        
        # Check if 'Date' and 'Title' columns exist in the DataFrame
        if 'Date' in df.columns and 'Title' in df.columns:
            # Select only 'Date' and 'Title' columns
            df_selected = df[['Date', 'Title']]
            
            # Clean the 'Title' column
            df_selected['Title'] = df_selected['Title'].apply(clean_title)
            
            # Append the selected data to the combined DataFrame
            df_combined = pd.concat([df_combined, df_selected], ignore_index=True)
        else:
            print(f"Warning: 'Date' or 'Title' column not found in {file.name}")

    # Convert 'Date' column to datetime, specifying the format
    df_combined['Date'] = pd.to_datetime(df_combined['Date'], format='%d-%m-%Y', errors='coerce')

    # Remove any rows where date conversion failed
    df_combined = df_combined.dropna(subset=['Date'])

    # Sort the DataFrame by date
    df_combined = df_combined.sort_values('Date')

    # Reset the index after sorting
    df_combined = df_combined.reset_index(drop=True)

    return df_combined

# Specify the directory path
directory = '../../Data/News/'

# Process the CSV files and get the combined DataFrame
df_combined = process_csv_files(directory)

# Display the first few rows of the combined DataFrame
print(df_combined.head())

# Display the date range in the data
print(f"Date range: from {df_combined['Date'].min()} to {df_combined['Date'].max()}")

# Optional: Display a sample of cleaned titles
print("\nSample of cleaned titles:")
print(df_combined['Title'].head())

# Save the combined DataFrame to a new CSV file (optional)
df_combined.to_csv('../../Data/News/combined_data_sorted.csv', index=False)
```
